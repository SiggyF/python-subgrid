```python
import os
import python_subgrid.wrapper
import ctypes
import logging
import io

from ctypes import c_char_p, POINTER, c_double, CFUNCTYPE, byref
```

```python
mdu = os.path.abspath('../../models/purmer/purmer.mdu')
```

```python
# subscribe to logging messages
stream = io.BytesIO()
handler = logging.StreamHandler(stream)
# remove all other handlers
python_subgrid.wrapper.progresslogger.parent = None
# set level to DEBUG so we get all the progress details
python_subgrid.wrapper.progresslogger.setLevel(logging.DEBUG)
python_subgrid.wrapper.progresslogger.addHandler(handler)
```

```python
subgrid = python_subgrid.wrapper.SubgridWrapper(mdu=mdu)
subgrid.start()
```
