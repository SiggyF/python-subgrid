```python
import python_subgrid.wrapper
import os.path
import osgeo.gdal
import netCDF4
import numpy as np
import matplotlib.pyplot as plt
```

```python
mdu = os.path.expanduser('~/Checkouts/models/delfland-model-voor-3di/hhdlipad.mdu')
mdudir = os.path.dirname(mdu)
#!head $mdu
tifffile = os.path.join(mdudir, 'subgrid/ahn_guts_v9.tif')
```

```python

tiff = osgeo.gdal.Open(tifffile)
```

```python
band = tiff.GetRasterBand(1)
```

```python
nodata = band.GetNoDataValue()
d = band.ReadAsArray()
dmasked = np.ma.masked_array(d, d==nodata)
plt.imshow(dmasked, cmap='gist_earth', vmin=-10, vmax=10)
```

```python
x0, dxx, dyx, y0, dyx, dyy= tiff.GetGeoTransform()


thin = 2

nx = tiff.RasterXSize
ny = tiff.RasterYSize

dxx*=thin
dyy*=thin
nx/=thin
ny/=thin

x = np.arange(x0, x0+dxx*nx, dxx)
y = np.arange(y0, y0+dyy*ny, dyy)
fig, ax = plt.subplots()
ax.pcolorfast(x,y, dmasked, cmap='gist_earth', vmin=-10, vmax=10)
```

```python
rain = np.zeros_like(dmasked[::thin,::thin])


rain[:,np.floor(nx/2):] = 200
rain.mask = np.logical_or(dmasked[::thin,::thin].mask, rain==0)
ax.pcolorfast(x,y, rain, alpha=0.6, cmap='Reds', vmin=0, vmax=200)
fig
```

```python
subgrid = python_subgrid.wrapper.SubgridWrapper(mdu=mdu)

subgrid.start()
```

```python
import logging
python_subgrid.wrapper.logger.setLevel(logging.INFO)
```

```python


print(nx, ny)
# Now store the variable in the model
ds = netCDF4.Dataset("precipitation.nc", mode="w", diskless=True, persist=False)

ds.createDimension("nx", nx)
ds.createDimension("ny", ny)

var = ds.createVariable("x", datatype="double", dimensions=("nx",))
var[:] = x[:]
var.standard_name = 'projected_x_coordinate'
var.units = 'm'

var = ds.createVariable("y", datatype="double", dimensions=("ny", ))
var[:] = y[:]
var.standard_name = 'projected_y_coordinate'
var.units = 'm'

var = ds.createVariable("rainfall", datatype="double", dimensions=("ny", "nx"), fill_value=rain.fill_value)
var[:,:] = rain.filled(0)
var.standard_name = 'precipitation'
var.coordinates = 'y x'
var.units = 'mm3/s'

s0 = subgrid.get_nd('s1').copy()
# Don't close the dataset before the model has done the timestep
subgrid.subscribe_dataset('precipitation.nc')
subgrid.update(10.0)
# Clears the memory
ds.close()
```
