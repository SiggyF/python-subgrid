```python
def make_quad_grid(grid):
    """create a quad grid based on the grid information"""
    # Create lookup index
    m = (grid['nodm']-1)*grid['dx'][grid['nodk']-1]/grid['dxp']
    n = (grid['nodn']-1)*grid['dx'][grid['nodk']-1]/grid['dxp']
    size = np.round(grid['dx'][grid['nodk']-1]/grid['dxp']).astype('int32')
    quad_grid = np.ma.empty(((n+size).max(), (m + size).max()), dtype='int32')
    quad_grid.mask = True
    for i, (m_i, n_i, size_i) in enumerate(zip(m,n,size)):
        quad_grid[n_i:(n_i+ size_i), m_i:(m_i+size_i)] = i
    return quad_grid

grid = {}
for var in {'nodm', 'nodn', 'nodk', 'nod_type', 'ip', 'jp', 'dx', 'dxp', 'dps', 'mmax', 'nmax', 'imax', 'jmax'}:
    grid[var] = subgrid.get_nd(var)
quad_grid = make_quad_grid(grid)
```

```python
m = (grid['nodm']-1)*grid['dx'][grid['nodk']-1]/grid['dxp']
n = (grid['nodn']-1)*grid['dx'][grid['nodk']-1]/grid['dxp']
size = np.round(grid['dx'][grid['nodk']-1]/grid['dxp']).astype('int32')
quad_grid = np.ma.empty(((n+size).max(), (m + size).max()), dtype='int32')
quad_grid.mask = True
for i, (m_i, n_i, size_i) in enumerate(zip(m,n,size)):
    quad_grid[n_i:(n_i+ size_i), m_i:(m_i+size_i)] = i
mc = m + size/2.0
nc = n + size/2.0
```

```python
fig, ax = plt.subplots(figsize=(20,13))
ax.imshow(np.ma.masked_less(grid['dps'], -10000), origin='lower')
ax.imshow(quad_grid, cmap='Set2', alpha=0.5, origin='lower')
ax.plot(mc, nc, 'k+', alpha=0.5)
```

```python

def C(x, cmap='Blues'):
    N = matplotlib.colors.Normalize(vmin=-5, vmax=0)
    cm = matplotlib.cm.cmap_d[cmap]
    colors = cm(N(x), alpha=1)  
    # add a row completely transparent
    colors = np.r_[colors, np.array([0,0,0,0])[np.newaxis,:]]
    return colors
```

```python

s1 = subgrid.get_nd('s1')[1:]
colors = C(s1, cmap='jet')

plt.imshow(colors[quad_grid.filled(colors.shape[0]-1)])
```

```python
import scipy.interpolate

N, M = np.mgrid[0:grid['dps'].shape[0], 0:grid['dps'].shape[1]]
fig, axes = plt.subplots(1, 2, figsize=(20,8), sharey=True, sharex=True)
axes[0].set_title('Non interpolated')
axes[0].imshow(C(s1, cmap='jet')[quad_grid.filled(-1)])
L = scipy.interpolate.LinearNDInterpolator(np.c_[mc, nc], s1)
axes[0].triplot(L.points[:,0], L.points[:,1], L.tri.vertices, alpha=0.2)
axes[1].set_title('Linear cell center')
axes[1].imshow(L(M, N))
axes[1].triplot(L.points[:,0], L.points[:,1], L.tri.vertices, alpha=0.1)

```
