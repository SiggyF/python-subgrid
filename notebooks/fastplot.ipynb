```python
import itertools
```

```python
def keyfunc(x):
    return tuple(x[:2])
```

```python
V = np.r_[
    np.c_[ll_x, ll_y, s1],
    np.c_[lr_x, lr_y, s1],
    np.c_[ul_x, ul_y, s1],
    np.c_[ur_x, ur_y, s1]
    ]
V = sorted(V, key=keyfunc)
```

```python
pts = []
means = []
for (key, value) in itertools.groupby(V, keyfunc):
    arr = array(list(value))
    pts.append(key)
    means.append(np.mean(arr[:,2]))
```

```python
plt.scatter(np.array(pts)[:,0], np.array(pts)[:,1], c=means, edgecolor='none')
plt.colorbar()
```
