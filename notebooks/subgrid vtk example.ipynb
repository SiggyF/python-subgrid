```python
import mayavi.mlab
```

```python
xc = subgrid.get_nd('FlowElemContour_x').copy()
yc = subgrid.get_nd('FlowElemContour_y').copy()
s1 = subgrid.get_nd('s1').copy()
```

```python
verts = [np.c_[x,y] for x,y in zip(xc.T, yc.T)]
```

```python
polys = matplotlib.collections.PolyCollection(verts)
fig, ax = plt.subplots(figsize=(20,15))
N = matplotlib.colors.Normalize(0, s1[1:].max())

colors= matplotlib.cm.Accent(N(s1[1:]))
polys.set_facecolors(colors)
ax.add_collection(polys)
ax.autoscale()
```

```python
from tvtk.api import  tvtk
```

```python
grid = tvtk.UnstructuredGrid()
ncells = xc.shape[1]
X = np.tile(xc.T, (1,2))
Y = np.tile(yc.T, (1,2))
Z = np.c_[ np.zeros(xc.T.shape)- (s1[1:].min() - 1), np.repeat(s1[1:,np.newaxis], 4, axis=1)]

pts = np.c_[X.ravel(), Y.ravel(), Z.ravel() ]
pts[:8]
```

```python

cell_array = tvtk.CellArray()
cell_types = np.array([tvtk.Hexahedron().cell_type for i in range(ncells)])
cell_idx = np.array([[8] + [i*8 + j for j in range(8)] for i in range(ncells)]).ravel()
cell_offsets = np.array([i*9 for i in range(ncells)])

cell_array.set_cells(ncells, cell_idx)
```

```python
grid.points = pts
grid.set_cells(cell_types, cell_offsets, cell_array)
grid.cell_data.scalars = s1[1:].copy()
grid.cell_data.scalars.name = 's1'
```

```python
import mayavi.sources.vtk_data_source
source = mayavi.sources.vtk_data_source.VTKDataSource(data=grid)
```

```python
engine = mayavi.mlab.get_engine()

outline = mayavi.modules.outline.Outline()
surface = mayavi.modules.surface.Surface()
```

```python
engine.add_source(source)
engine.add_module(outline)
engine.add_module(surface)
mayavi.mlab.show()
```
